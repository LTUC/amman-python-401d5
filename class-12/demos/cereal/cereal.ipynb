# Show the food name and sugar content of any food with a rating larger than 50

```python
highly_rated_food =  df[df['rating'] > 50][['name', 'sugars']]
highly_rated_food.set_index('name')
```

```python
mcc =  df['carbo'].mode()
mcc
```

```python
import numpy as np
df[df['carbo'] == np.mean(mcc)][['name', 'carbo']].set_index('name')
```

## What is the median value for protein and sugars across / for/  per each manufacturer?

```python
df[['mfr','sugars','protein']].groupby('mfr').median()
```

```python
def manufacturer_names(mfr):
    map = {
        "A": "Alfalfa",
        "G": 'General Mills',
        "K": 'Kellogs',
        "N": 'Nabisco',
        "P": 'Post',
        "Q": 'Quaker',
        "R": 'Ralcorp',
    }
    return map.get(mfr)

df['Manufacturer'] = df['mfr'].apply(manufacturer_names)
df
```
